```python
import numpy as np

# Coefficient matrix A
A = np.array([
    [1, 2, 1],
    [2, 3, 1],
    [-1, 1, -2]
])

# Constants vector B
B = np.array([9, 16, -1])

# Solving Ax = B
solution = np.linalg.solve(A, B)

# Print the result
x, y, z = solution
print(f"x = {x}")
print(f"y = {y}")
print(f"z = {z}")

```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Your solution point
x = 4.5
y = 2.5
z = -0.5

# Plot the point
ax.scatter(x, y, z, color='red', s=100, label='Solution (4.5, 2.5, -0.5)')

# Set axis labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Set title and legend
ax.set_title('3D Solution Point')
ax.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x and y values
x = np.linspace(-10, 10, 10)
y = np.linspace(-10, 10, 10)
X, Y = np.meshgrid(x, y)

# Define the equations of planes
# 1. x + 2y + z = 9  => z = 9 - x - 2y
Z1 = 9 - X - 2*Y

# 2. 2x + 3y + z = 16 => z = 16 - 2x - 3y
Z2 = 16 - 2*X - 3*Y

# 3. -x + y - 2z = -1 => z = (1 - x + y)/2
Z3 = (1 - (-X + Y)) / 2  # Simplified

# Plotting
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the planes
ax.plot_surface(X, Y, Z1, color='red', alpha=0.6, label='Plane 1')
ax.plot_surface(X, Y, Z2, color='blue', alpha=0.6, label='Plane 2')
ax.plot_surface(X, Y, Z3, color='purple', alpha=0.6, label='Plane 3')

# Label axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Set the title
ax.set_title('3D Plot of the Planes')

plt.show()
```

## Solving Complex Equations

```python
import numpy as np

# Coefficients of the quadratic equation
a = 1
b = 3 - 4j
c = 7 + 24j

# Calculate the discriminant
discriminant = np.sqrt(b**2 - 4*a*c)

# Apply the quadratic formula
z1 = (-b + discriminant) / (2*a)
z2 = (-b - discriminant) / (2*a)

# Print the solutions
print("Solution 1: ", z1)
print("Solution 2: ", z2)
```

## Vectors in Python

```python
v = np.array([2, 3])
```

```python
v
```

```python
# plt.figure(figsize=(8, 6))
plt.plot([0, v[0]], [0, v[1]], 'r->', linewidth=2, label='Vector v')
plt.plot([0], [0], 'ko', label='Origin') 
plt.grid(True)
plt.axis('equal')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.title(f'Vector v = [{v[0]}, {v[1]}]')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.tight_layout()
```

```python
# Perform linear transformation using matrix A
# We need to pad v with a third component (0) since A is 3x3
v_padded = np.append(v, 0)
transformed_v = A @ v_padded

# Create a 2D plot to show both original and transformed vectors
plt.figure(figsize=(8,6))
# Plot original vector
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original vector')
# Plot transformed vector (using only first 2 components)
plt.quiver(0, 0, transformed_v[0], transformed_v[1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed vector')

plt.grid(True)
plt.axis('equal')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.title('Linear Transformation of Vector v')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Move the original vector v by rotating it 45 degrees counterclockwise
theta = np.deg2rad(50)
rotation_matrix = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])
v_rotated = rotation_matrix @ v

# Plot original, transformed, and rotated vectors
plt.figure(figsize=(8,6))
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original vector')
plt.quiver(0, 0, transformed_v[0], transformed_v[1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed vector')
plt.quiver(0, 0, v_rotated[0], v_rotated[1], angles='xy', scale_units='xy', scale=1, color='green', label='Rotated vector (45°)')

plt.grid(True)
plt.axis('equal')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.title('Linear Transformation and Rotation of Vector v')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Calculate vector norms and angles
magnitude_v = np.linalg.norm(v)
angle_rotated = np.arccos(np.dot(v, v_rotated) / (np.linalg.norm(v) * np.linalg.norm(v_rotated)))
angle_degrees = np.rad2deg(angle_rotated)

# Create a polar plot to show the vectors and angles
plt.figure(figsize=(5, 5))
ax = plt.subplot(111, projection='polar')

# Plot vectors in polar coordinates
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original vector')
ax.quiver(0, 0, v_rotated[0], v_rotated[1], angles='xy', scale_units='xy', scale=1, color='green', label='Rotated vector')

# Add annotations
plt.title(f'Vector Rotation Analysis\nRotation Angle: {angle_degrees:.2f}°\nVector Magnitude: {magnitude_v:.2f}')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Create a grid of 2D points
grid_x, grid_y = np.meshgrid(np.linspace(-5, 5, 11), np.linspace(-5, 5, 11))
points = np.vstack([grid_x.ravel(), grid_y.ravel()])

# Define a shear transformation matrix (shear in x-direction)
shear_factor = 1.0
shear_matrix = np.array([
    [1, shear_factor],
    [0, 1]
])

# Apply the shear transformation
sheared_points = shear_matrix @ points

# Plot original and sheared grids
plt.figure(figsize=(5, 5))
plt.plot(points[0], points[1], 'bo', label='Original grid')
plt.plot(sheared_points[0], sheared_points[1], 'ro', label='Sheared grid')

# Draw grid lines for better visualization
for i in range(points.shape[1]):
    plt.plot([points[0, i], sheared_points[0, i]], [points[1, i], sheared_points[1, i]], 'k--', alpha=0.3)

plt.title('2D Shear Transformation of Grid')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

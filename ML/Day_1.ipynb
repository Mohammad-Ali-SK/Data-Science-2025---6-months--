```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Set plot style
sns.set(style="whitegrid")

# Create the plot with regression line
plt.figure(figsize=(10, 6))
ax = sns.regplot(
    x='total_bill',
    y='tip',
    data=df,
    scatter_kws={'color': 'skyblue', 's': 60, 'edgecolor': 'w'},
    line_kws={'color': 'red', 'linewidth': 2}
)

# Annotate only points where total_bill > 40
for i in range(len(df)):
    x = df['total_bill'].iloc[i]
    y = df['tip'].iloc[i]
    if x > 40:
        ax.text(x, y + 0.3, f"({x:.1f}, {y:.1f})", fontsize=8, color='darkblue')

# Add labels and title
plt.title('Total Bill vs Tip (Labeling High Bills Only)', fontsize=14)
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Tip ($)', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#selecting features columns
x = df[['total_bill']]

#selecting target variable

y = df['tip']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#load the model
model = LinearRegression()
```

```python
# Fit th model
model.fit(x,y)
```

```python
# Lets predict the tip for a total bill of 50
total_bill = np.array([[16.99]])
predicted_tip = model.predict(total_bill)
```

```python
tips = predicted_tip.tolist()[0]
print(f"Predicted tip for a total bill of ${total_bill.tolist()[0][0]}: ${tips:.2f}")
```

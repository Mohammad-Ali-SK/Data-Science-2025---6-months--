```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PowerTransformer, QuantileTransformer

# Generate synthetic skewed data
np.random.seed(42)
data = np.random.exponential(scale=2, size=500).reshape(-1, 1)  # right-skewed

# Apply Power Transformation (Yeo-Johnson)
pt = PowerTransformer(method='yeo-johnson')
data_power = pt.fit_transform(data)

# Apply Quantile Transformation (to normal distribution)
qt = QuantileTransformer(output_distribution='normal', random_state=42)
data_quantile = qt.fit_transform(data)

# Plot the distributions
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Original data
axes[0].hist(data, bins=30, color='skyblue', edgecolor='black')
axes[0].set_title("Original Skewed Data")

# Power transformed data
axes[1].hist(data_power, bins=30, color='lightgreen', edgecolor='black')
axes[1].set_title("Power Transformation (Yeo-Johnson)")

# Quantile transformed data
axes[2].hist(data_quantile, bins=30, color='salmon', edgecolor='black')
axes[2].set_title("Quantile Transformation (Normal Dist.)")

plt.tight_layout()
plt.show()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PowerTransformer, QuantileTransformer

# Generate synthetic skewed data
np.random.seed(42)
data = np.random.exponential(scale=2, size=500).reshape(-1, 1)  # right-skewed

# Apply Power Transformation (Yeo-Johnson)
pt = PowerTransformer(method='yeo-johnson')
data_power = pt.fit_transform(data)

# Apply Quantile Transformation (to normal distribution)
qt = QuantileTransformer(output_distribution='normal', random_state=42)
data_quantile = qt.fit_transform(data)

# Plot the distributions
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Original data
axes[0].hist(data, bins=30, color='skyblue', edgecolor='black')
axes[0].set_title("Original Skewed Data")

# Power transformed data
axes[1].hist(data_power, bins=30, color='lightgreen', edgecolor='black')
axes[1].set_title("Power Transformation (Yeo-Johnson)")

# Quantile transformed data
axes[2].hist(data_quantile, bins=30, color='salmon', edgecolor='black')
axes[2].set_title("Quantile Transformation (Normal Dist.)")

plt.tight_layout()
plt.show()
```

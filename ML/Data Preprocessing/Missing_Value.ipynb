# 2. K- Nearest Neighbors (KNN) Algorithm

```python
# Load important Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load a dataset
data = sns.load_dataset('titanic')
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print(f"Missing values in each column:\n {missing_values}")
```

```python
# Impute missing values with KNN Imputer
from sklearn.impute import KNNImputer

# call the KNNImputer
imputer = KNNImputer(n_neighbors=10)

# input missing values with KNN
data['age'] = imputer.fit_transform(data[['age']])


# check the missing values after imputation

data.isnull().sum()
```

```python
data['age'].mean()
```

```python
data.isnull().sum().sort_values(ascending=False)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Load the dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
# Check for missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
# Import Label Encoder
from sklearn.preprocessing import LabelEncoder

# Columns to encode
columns_to_encode = ['sex', 'embarked','class', 'who', 'deck', 'embark_town', 'alive']

# Dictionary to hold encoders
label_encoders = {}

# Loop through each column and apply Label Encoding
for column in columns_to_encode:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column].astype(str))  # Convert to string to avoid errors with NaN
    label_encoders[column] = le
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
imputer = KNNImputer(n_neighbors=10)

df['age'] = imputer.fit_transform(df[['age']])
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
for col in label_encoders:
    print(label_encoders[col])
```

## Inverse Transform for encoded columns

```python
for col in columns_to_encode:
    le = label_encoders[col]
    df[col] = le.inverse_transform(df[col].astype(int))
df.head()
```

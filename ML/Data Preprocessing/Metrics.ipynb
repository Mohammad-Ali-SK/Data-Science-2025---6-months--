## üìò Linear Regression vs Logistic Regression ‚Äì Including Metrics

### üîπ What They Have in Common:

* Both are **supervised learning algorithms**
* Both can use similar input features
* Both estimate **coefficients (weights)** for features

---

## üî∑ Linear Regression

### ‚úÖ Used For:

* **Predicting continuous values**

### üìà Equation:

$$
y = b_0 + b_1x_1 + b_2x_2 + ... + b_nx_n
$$

### üîß Example:

Predict house price based on area.

### üìè Evaluation Metrics:

| Metric   | Description                              |
| -------- | ---------------------------------------- |
| MAE      | Mean Absolute Error                      |
| MSE      | Mean Squared Error                       |
| RMSE     | Root Mean Squared Error                  |
| R¬≤ Score | Variance explained by the model (0 to 1) |

---

## üî∂ Logistic Regression

### ‚úÖ Used For:

* **Binary or multiclass classification**

### üìà Equation:

Uses **sigmoid function**:

$$
p = \frac{1}{1 + e^{-z}}, \quad \text{where } z = b_0 + b_1x_1 + b_2x_2 + ... + b_nx_n
$$

### üîß Example:

Predict whether a customer will churn (Yes/No)

### üìè Evaluation Metrics:

| Metric    | Description                                               |
| --------- | --------------------------------------------------------- |
| Accuracy  | % of correct predictions                                  |
| Precision | % of predicted positives that are correct                 |
| Recall    | % of actual positives identified correctly                |
| F1 Score  | Harmonic mean of Precision and Recall                     |
| ROC AUC   | Area under ROC Curve; model‚Äôs ability to separate classes |

---

## üîÅ Comparison Table

| Feature            | Linear Regression        | Logistic Regression                   |
| ------------------ | ------------------------ | ------------------------------------- |
| Type               | Regression               | Classification                        |
| Output             | Continuous (real values) | Probability (0 to 1) ‚Üí Class (0 or 1) |
| Equation           | Straight line            | Sigmoid/logistic function             |
| Output Type        | Numeric                  | Categorical (binary or multiclass)    |
| Loss Function      | MSE (Mean Squared Error) | Log Loss (Cross-Entropy)              |
| Evaluation Metrics | MAE, MSE, RMSE, R¬≤       | Accuracy, Precision, Recall, F1, AUC  |

---

‚úÖ Both models are essential: Linear Regression is your tool for predicting **how much**, and Logistic Regression is your tool for predicting **yes/no** decisions.
